```python
df = pd.merge(ratings, anime, on="anime_id")
# df.head()
```

```python
df.head()
```

```python
df=df.drop(['genre','rating_y','members'],axis=1)
```

```python
rating_count=(df.groupby(by=['name'])['rating_x'].count().reset_index().rename(columns={'rating_x':'rating_count'})[['name','rating_count']])
rating_count.head()
```

```python
print(rating_count.columns)
```

```python
# Clean the 'name' column in the rating_count dataframe
rating_count['name'] = rating_count['name'].str.replace('&quot;', '', regex=False)  # Remove HTML entities
rating_count['name'] = rating_count['name'].str.replace(r'[^\w\s]', '', regex=True)  # Remove unwanted characters
rating_count['name'] = rating_count['name'].str.strip()  # Remove leading/trailing whitespace
rating_count['name'] = rating_count['name'].str.lower()  # Convert to lowercase (optional)

# Display the cleaned dataframe
print(rating_count)

df['name'] = df['name'].str.replace('&quot;', '', regex=False)  # Remove HTML entities
df['name'] = df['name'].str.replace(r'[^\w\s]', '', regex=True)  # Remove unwanted characters
df['name'] = df['name'].str.strip()  # Remove leading/trailing whitespace
df['name'] = df['name'].str.lower()
```

```python
combined_rating_count=df.merge(rating_count, left_on = 'name', right_on = 'name', how = 'left')
combined_rating_count.head()
```

```python
print(rating_count['rating_count'].describe())
```

```python
print(rating_count['rating_count'].quantile(np.arange(.9, 1, .01)))
```

```python
popularity_threshold = 7000
popular_ratings= combined_rating_count.query('rating_count >= @popularity_threshold')
popular_ratings.tail()
popular_ratings.info()
```

```python
popular_ratings
```

```python
popular_ratings = popular_ratings.drop_duplicates(['user_id', 'name'])
print(popular_ratings.shape)
```

```python
pivottable=popular_ratings.pivot(index='user_id',columns='name',values='rating_x')
pivottable.shape
```

```python
pivottable.head()
```

```python
pivottable=pivottable.fillna(0)
pivottable.head()
```

```python
X=pivottable.values.T
X.shape
```

```python
from sklearn.decomposition import TruncatedSVD
from sklearn.feature_extraction.text import TfidfVectorizer
SVD= TruncatedSVD(n_components=12,random_state=17)
matrix=SVD.fit_transform(X)
matrix.shape
```

```python
# import warnings
# warnings.filterwarnings("ignore",category=RuntimeWarning)
corr=np.corrcoef(matrix)
corr.shape
```
